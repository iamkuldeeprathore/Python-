```python
import logging 

logger=logging.getLogger(__name__) #this for adding your custom logger

fh=logging.FileHandler('abc.log')
sh=logging.StreamHandler()

fh.setLevel(logging.WARNING)
sh.setLevel(logging.INFO)


#creating formatter and adding to handler
fh_format=logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
sh_format=logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')

fh.setFormatter(fh_format)
sh.setFormatter(sh_format)

```

```python
logger.addHandler(fh)
logger.addHandler(sh)
```
