```python
def test1(func):
    def test2(*args,**kwargs):
        print("this is start of test2 function")
        # func(*args, **kwargs)
        print("this is end of test2 fucntion")
        return func(*args,**kwargs)
    return test2
    
```

```python
@test1
def test3(a,b):
    return a+b
```

```python
test3(6,7)
```
