```python
def length(*args):
    count=0
    for ele in args[0]:
        print(ele)
        count+=1
    return count
    
```

```python
length("kuldeep")
```

```python
def indx(*args):
    i=0
    for ele in args[0]:
        if(ele==args[1]):
            return i
        else: i+=1
    return -1
           
            
            
                      
            
```

```python
import socket

try:
    # Get the hostname of the local machine
    hostname = socket.gethostname()

    # Get the IP address associated with the hostname
    ip_address = socket.gethostbyname(hostname)

    print(f"Your computer's hostname: {hostname}")
    print(f"Your computer's IP address: {ip_address}")

except socket.gaierror as e:
    print(f"Could not resolve hostname or IP address: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
import time

def executiontimng(prblm):
    def inn():
        st=time.time()
        prblm(67,9)
        en=time.time()
        print(f"Executation time for this function is {en-st}")
    return inn
    
```

```python

@executiontiming
def prblm(a,b):
    print(a+b)
    
```
