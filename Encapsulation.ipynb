```python
class Car:
    
    def __init__(self,car_company,car_model,color):
        self.car_company=car_company
        self.car_model=car_model
        self.color=color

    def show(self):
        print("Company : ",self.car_company,"\nCar model :",self.car_model,"\nColor :",self.color)
        
        
    def __str__(self):
        return ("this is car class")
```

```python
car2=Car("Tata","nano","yellow")
```

```python
class Car:
    
    def __init__(self,car_company,car_model,color):
        self._car_company=car_company
        self._car_model=car_model
        self._color=color

    def show(self):
        print("Company : ",self._car_company,"\nCar model :",self._car_model,"\nColor :",self._color)
        
        
    def __str__(self):
        return ("this is car class")
```

```python
car3=Car("honda","suv","black")
```

```python
car3.show()
```

```python
car3.car_company
```

```python
# can't access the protected variable but can use in class or subclass
```

```python
car3._car_model
```

```python
class Car:
    
    def __init__(self,car_company,car_model,color):
        self.__car_company=car_company
        self.__car_model=car_model
        self.__color=color

    def show(self):
        print("Company : ",self.__car_company,"\nCar model :",self.__car_model,"\nColor :",self.__color)
        
        
    def __str__(self):
        return ("this is car class")
```

```python
car4=Car("tata","nano","white")
```

```python
car4.car_company
```

```python
car4._car_model
```
