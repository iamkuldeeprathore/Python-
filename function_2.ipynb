```python
# filter function
```

```python
def test4(a):
    if a%2==0:
        return True
    else:
        return False
```

```python
lii=[1,2,4,3,5,6,7,7,5,5,]
```

```python
list(filter(test4,lii))
```

```python
list(filter(lambda a :a%2==0,lii))
```
