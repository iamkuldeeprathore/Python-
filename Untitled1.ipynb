```python
class Stack:

    def __init__(self):          # âœ… FIXED
        self.stack = []

    def add(self, val):
        self.stack.append(val)

    def pop(self):
        if not self.isEmpty():
            return self.stack.pop()
        else:
            print("Underflow condition")

    def isEmpty(self):
        return len(self.stack) == 0

    def size(self):
        return len(self.stack)

    def top(self):
        if not self.isEmpty():
            return self.stack[-1]
        return -1
        
    def display(self):
        if self.isEmpty():
            print("Stack is empty")
        else:
            for i in self.stack:
                print(i, end=" ")
```
