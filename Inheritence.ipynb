```python
class Person1:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print("Name:", self.name, "Age:", self.age)


class Person2:
    def __init__(self, lname, weight):
        self.lname = lname
        self.weight = weight

    def display1(self):
        print("Last Name:", self.lname, "Weight:", self.weight)


class Child(Person1, Person2):
    def __init__(self, name, age, lname, weight):
        Person1.__init__(self, name, age)  # calling first parent
        Person2.__init__(self, lname, weight)  # calling second parent
        print("I am in child constructor")


c = Child("Kuldeep", 25, "Rathore", 70)
c.display()      # From Person1
c.display1()     # From Person2
```

```python
class Vehicle:
    def __init__(self,name,model):
        self.name=name
        self.model=model

    def display(self):
        print("vehicle name ",self.name,"model :",self.model)
        
class Car(Vehicle):
    def __init__(self,car_name):
        self.car_name=car_name
        
    def show(self):
        print("car name is :",self.car_name)

class Truck(Car):
    def __init__(self,truck_name):
        self.truck_name=truck_name
        Vehicle.__init__(self,name,model)
        Car.__init__(self,car_name)
        
    def show1():
        print("truck name :",self.truck_name)
```

```python
class Vehicle:
    def __init__(self, name, model):
        self.name = name
        self.model = model

    def display(self):
        print("Vehicle Name:", self.name, "Model:", self.model)


class Car(Vehicle):
    def __init__(self, name, model, car_name):
        super().__init__(name, model)  # calls Vehicle constructor
        self.car_name = car_name

    def show(self):
        print("Car Name:", self.car_name)


class Truck(Car):
    def __init__(self, name, model, car_name, truck_name):
        super().__init__(name, model, car_name)  # calls Car → Vehicle
        self.truck_name = truck_name

    def show1(self):
        print("Truck Name:", self.truck_name)


# Object Creation
t = Truck("Tata", "2021", "Safari", "Tata Truck")
t.display()
t.show()
t.show1()
```

```python
class A:
    def __init__(self, a):
        self.a = a

    def showA(self):
        print("A =", self.a)


class B(A):
    def __init__(self, a, b):
        super().__init__(a)
        self.b = b

    def showB(self):
        print("B =", self.b)


class C(B):
    def __init__(self, a, b, c):
        super().__init__(a, b)
        self.c = c

    def showC(self):
        print("C =", self.c)


obj = C(10, 20, 30)
obj.showA()
obj.showB()
obj.showC()
```

```python
class Father:
    def __init__(self, fname):
        self.fname = fname

    def showF(self):
        print("Father Name:", self.fname)


class Mother:
    def __init__(self, mname):
        self.mname = mname

    def showM(self):
        print("Mother Name:", self.mname)


class Child(Father, Mother):
    def __init__(self, fname, mname, cname):
        Father.__init__(self, fname)
        Mother.__init__(self, mname)
        self.cname = cname

    def showC(self):
        print("Child Name:", self.cname)


obj = Child("Raj", "Rani", "Raju")
obj.showF()
obj.showM()
obj.showC()

```

```python
class Parent:
    def __init__(self, name):
        self.name = name

    def show(self):
        print("Parent Name:", self.name)


class Child1(Parent):
    def __init__(self, name, grade):
        super().__init__(name)
        self.grade = grade

    def display1(self):
        print("Child1:", self.name, "Grade:", self.grade)


class Child2(Parent):
    def __init__(self, name, marks):
        super().__init__(name)
        self.marks = marks

    def display2(self):
        print("Child2:", self.name, "Marks:", self.marks)


c1 = Child1("Deepak", "A")
c2 = Child2("Mohit", 92)
c1.show()
c1.display1()
c2.show()
c2.display2()
```

```python
class A:
    def __init__(self, a):
        self.a = a

    def showA(self):
        print("A =", self.a)


class B(A):
    def __init__(self, a, b):
        super().__init__(a)
        self.b = b

    def showB(self):
        print("B =", self.b)


class C(A):
    def __init__(self, a, c):
        super().__init__(a)
        self.c = c

    def showC(self):
        print("C =", self.c)


class D(B, C):  # inherits from both B and C
    def __init__(self, a, b, c, d):
        B.__init__(self, a, b)
        C.__init__(self, a, c)
        self.d = d

    def showD(self):
        print("D =", self.d)


obj = D(10, 20, 30, 40)
obj.showA()
obj.showB()
obj.showC()
obj.showD()
```

```python
class A:
    def __init__(self, a):
        self.a = a

    def showA(self):
        print("A =", self.a)


class B(A):
    def __init__(self, a, b):
        A.__init__(self, a)
        self.b = b

    def showB(self):
        print("B =", self.b)


class C(A):
    def __init__(self, a, c):
        A.__init__(self, a)
        self.c = c

    def showC(self):
        print("C =", self.c)


class D(B, C):  # Hybrid Inheritance
    def __init__(self, a, b, c, d):
        B.__init__(self, a, b)   # call B constructor
        C.__init__(self, a, c)   # call C constructor
        self.d = d

    def showD(self):
        print("D =", self.d)


obj = D(10, 20, 30, 40)
obj.showA()
obj.showB()
obj.showC()
obj.showD()
```
